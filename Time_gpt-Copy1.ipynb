```python
import pandas as pd
import numpy as np
from nixtla import NixtlaClient
```

## Step 2: Load and Preprocess the Data

```python
# Load the dataset
file_path = "C:/Users/shaan/Downloads/AirPassengers.csv"
df = pd.read_csv(file_path)
df.head()
```

## Step 3: Initialize NixtlaClient and Validate API Key

```python
nixtla_client = NixtlaClient(
    api_key = '******'
)
nixtla_client.validate_api_key()
```

## Step 4: Visualize Historical Data

```python
nixtla_client.plot(df, time_col='Month', target_col='#Passengers')
```

## Step 5: Generate Forecast Using TimeGPT

```python
timegpt_fcst_df = nixtla_client.forecast(df=df, h=12, freq='MS', time_col='Month', target_col='#Passengers')
timegpt_fcst_df.head()
```

```python
nixtla_client.plot(df, timegpt_fcst_df, time_col='Month', target_col='#Passengers')
```

```python
timegpt_fcst_df = nixtla_client.forecast(df=df, h=36, time_col='Month', target_col='#Passengers', freq='MS', model='timegpt-1-long-horizon')
timegpt_fcst_df.head()
```

```python
nixtla_client.plot(df, timegpt_fcst_df, time_col='Month', target_col='#Passengers')
```
